- Representing the marginal probability like what percent of customers purchased KP281, KP781 OR KP481 in a table can use pandas.crosstab

```python
# Create a crosstab table of the product ID and the number of customers who purchased each product
crosstab_table = pd.crosstab(index= dftm["Product"], columns=dftm["Marital Status"])

```

```python
crosstab_table
```

```python
pd.crosstab(index = [dftm["Product"], dftm["Gender"]], columns = dftm["Marital Status"], values = dftm["Age"], aggfunc = "mean", normalize = True, margins = True)
```

```python
crosstab_table = pd.crosstab(index=dftm['Product'], columns='count')

marginal_probs = crosstab_table['count'] / crosstab_table['count'].sum()

marginal_probs_percentage = marginal_probs * 100

result_table = pd.DataFrame({'Marginal Probability': marginal_probs_percentage})

print(result_table)
```

- what is the probability of male customer buying KP781 treadmill?

```python
male_customers = dftm[dftm['Gender'] == 'Male']
total_male_customers = len(male_customers)

male_kp781_purchases = male_customers[male_customers['Product'] == 'KP781']
num_male_kp781_purchases = len(male_kp781_purchases)

probability_male_kp781 = num_male_kp781_purchases / total_male_customers

print("Probability of a male customer buying KP781 treadmill:", probability_male_kp781)
```
